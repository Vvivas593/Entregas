```python
import pandas as pd
```

### Step 3. Assign it to a variable called users.

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
users 
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby(by="occupation")[["age"]].mean().add_prefix("MeanOf_").reset_index()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
liste1 = users.groupby(by=["occupation","gender"])[["gender"]].count().add_prefix("CountOf_").reset_index()
liste2 = liste1.pivot(index="occupation",columns="gender",values="CountOf_gender")
liste2 = liste2.fillna(0)
```

```python
otre1 = liste2.F + liste2.M
otre2 = (liste2.M/otre1)*100
final_res = otre2.sort_values(ascending=False)
final_res
```

```python
ratios = []
occus = []

for occu in users.loc[:, "occupation"].unique():
    # print(occu)
    subset = users.loc[users["occupation"]==occu,:]
    # print(subset)
    male = subset.loc[subset["gender"] == "M"].shape[0]
    female = subset.loc[subset["gender"] == "F"].shape[0]
    ratio = (male/(male+female))*100
    occus.append(occu)
    ratios.append(ratio)

pd.DataFrame({"occupation": occus, "ratio":ratios}).sort_values("ratio", ascending=False)
```

```python
proportions = users.groupby("occupation")[["gender"]].value_counts(normalize=True).reset_index()
```

```python
proportions.loc[proportions.loc[:, "gender"]=="M", :].sort_values("proportion", ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
lista1 = users.groupby(by="occupation")[["age"]].min().add_prefix("Min_")
lista2 = users.groupby(by="occupation")[["age"]].max().add_prefix("Max_")
respuesta = pd.concat([lista1,lista2],axis=1)
respuesta
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
respueste = users.groupby(by=["occupation","gender"])[["age"]].mean().add_prefix("MeanOf_").reset_index()
respuesti = respueste.pivot(index="occupation",columns="gender",values="MeanOf_age")
respuesti
```
