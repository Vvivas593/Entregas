3. Sabemos que podemos pintar gr치ficas de dos formas: la figura incluye los ejes o teniendo figura y ejes por separado.

Usando solo una figura, usa numpy para los valores del eje X entre 0 y 5. 

Pinta dos gr치ficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). 

Elige la precisi칩n en el eje X que desees.  

```python
x = np.linspace(0, 5, 100)

y1 = 3 * x
y2 = -3 * x - 5

fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(x, y1)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
 
ax2.plot(x, y2)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

plt.show()
```

8. Crea un gr치fico de barras. Pintamos la altura de un grupo de amigos, cada barra representa a una persona. Ana mide 160 cm, Luis mide 180 cm, Pedro mide 175 cm, Sof칤a mide 190 cm, Carmen mide 170 cm. Las barras ser치n verdes.
Consejo: si no vas a tener que manipular especialmente los ejes, es m치s sencillo dejarlos dentro de la figura.

```python
import matplotlib.pyplot as plt

alturas = [160, 180, 175, 190, 170]

nombres = ['Ana', 'Luis', 'Pedro', 'Sof칤a', 'Carmen']

plt.bar(nombres, alturas, color='green')

plt.xlabel('Amigos')
plt.ylabel('Altura (cm)')
plt.title('Alturas de los Amigos')

plt.show()
```

9. Bas치ndote en el gr치fico anterior, escribe encima de cada barra la altura de cada amigo.
Pista: usa un bucle que lea cada barra de barplot = plt.bar(x,y)
bar tiene los m칠todos get_height(), get_x(), get_width()
plt.text(x,y,valor, va='bottom')

```python

barplot = plt.bar(nombres, alturas, color='green')

for bar in barplot:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, height, ha='center', va='bottom')

plt.xlabel('Amigos')
plt.ylabel('Altura (cm)')
plt.title('Alturas de los Amigos')

plt.show()


```

16. Ahora pintamos un gr치fico de l칤nea con las ventas respecto al a침o. Pon un t칤tulo y etiquetas en los ejes.
Cambia el estilo a 'seaborn-white'. La l칤nea debe ser con rayas y verde.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Year': [2015, 2016, 2017, 2018, 2019],
    'Sold_items_A': [1000, 3500, 4000, 5500, 7000]
}

df = pd.DataFrame(data)

df.plot(x='Year', y='Sold_items_A', kind='line', color='green', linestyle='--')

plt.title('Ventas respecto al a침o')
plt.xlabel('A침o')
plt.ylabel('Ventas')

plt.show()
```

```python

data = {
    'Year': [2015, 2016, 2017, 2018, 2019],
    'Sold_items_A': [1000, 3500, 4000, 5500, 7000],
    'Sold_items_B': [2000, 3100, 5000, 4000, 6000]
}

df = pd.DataFrame(data)
df.plot(x='Year', y='Sold_items_A', kind='line', color='green', linestyle='--', label='Departamento A')
df.plot(x='Year', y='Sold_items_B', kind='line', color='red', linestyle=':', label='Departamento B')

plt.title('Ventas respecto al a침o')
plt.xlabel('A침o')
plt.ylabel('Ventas')
plt.legend(loc='lower right')

plt.show()
```
