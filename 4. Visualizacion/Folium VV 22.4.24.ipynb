```python
import pandas as pd
import folium
import geojson

ruta = pd.DataFrame({'city': ['Casa', 'YogaBru', 'Playa', 'Cafe', 'Edem'],
                     'latitude': [39.4653687079709, 39.46543650729194, 39.465427423177424, 39.464919685421435, 39.46220028634176],
                     'longitude': [-0.332603704322556, -0.32848451581090793, -0.3236365682963254, -0.3298746913265121, -0.3290978465976334],
                     'icon': ['home', 'book', 'tree', 'coffee', 'university']})

features = []
for i, row in ruta.iterrows():
    feature = geojson.Feature(geometry=geojson.Point((row['longitude'], row['latitude'])),
                               properties={"city": row['city'], "icon": row['icon']})
    features.append(feature)

feature_collection = geojson.FeatureCollection(features)

with open("ruta.geojson", "w") as f:
    geojson.dump(feature_collection, f)

print("GeoJSON file 'ruta.geojson' created successfully.")

map_route
```
